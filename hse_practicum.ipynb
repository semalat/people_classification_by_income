```python
X = data.drop('income', axis=1) # матрица объект-признак
y = data['income'] # целевая переменная
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

numerical = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']

ss = MinMaxScaler()
ss.fit(Xtrain[numerical])

Xtrain_num = pd.DataFrame(ss.transform(Xtrain[numerical]), columns = numerical)
Xtest_num = pd.DataFrame(ss.transform(Xtest[numerical]), columns = numerical)

Xtrain_num.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical = ['workclass','education','marital-status','occupation', 'relationship', 'race', 'gender', 'native-country']

OH_encoder = OneHotEncoder(handle_unknown='error', sparse=False)
OH_encoder.fit(Xtrain[categorical])

Xtrain_ohe = pd.DataFrame(OH_encoder.transform(Xtrain[categorical]), columns=OH_encoder.get_feature_names_out())
Xtest_ohe = pd.DataFrame(OH_encoder.transform(Xtest[categorical]), columns=OH_encoder.get_feature_names_out())
```

```python
Xtrain_new = pd.concat([Xtrain_num, Xtrain_ohe], axis=1)
Xtest_new = pd.concat([Xtest_num, Xtest_ohe], axis=1)
```

```python
Xtrain_new[:10]
```

```python
from catboost import CatBoostClassifier

ytrain = ytrain.astype(int)

gbm = CatBoostClassifier(iterations=300, #сколько деревьев
                          depth=2, #глубина каждого дерева
                          learning_rate=0.1,
                          loss_function='Logloss',
                          eval_metric='AUC',
                          verbose=False)
gbm.fit(Xtrain_new, ytrain)
gbm_preds = gbm.predict(Xtest_new)
```

```python
ytest = ytest.astype(int)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(ytest, gbm_preds)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression() # объявление модели
# обучение модели
```

```python
model.fit(Xtrain_new, ytrain) 
```

```python
pred_classes = model.predict(Xtest_new) # предсказание классов
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest, pred_classes)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(ytest, pred_classes)
```

```python
from sklearn.model_selection import GridSearchCV

gbm2 = CatBoostClassifier(iterations = 500,
                          loss_function = 'Logloss',
                          learning_rate = 0.05,
                          eval_metric='AUC',
                          verbose=False)

params = {'max_depth' : [2, 3, 4, 5, 6]}

gs = GridSearchCV(gbm2, params, cv = 5, scoring='roc_auc')
```
